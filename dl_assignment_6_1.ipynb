```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense

n_days = 365 * 10
dates = pd.date_range(start='2010-01-01', periods=n_days, freq='D')

base_consumption = 1000


yearly_seasonality = 300 * np.sin(np.linspace(0, 2 * np.pi * 10, n_days)) + 200 * np.cos(np.linspace(0, 4 * np.pi * 10, n_days))


weekly_seasonality = 50 * np.sin(np.linspace(0, 2 * np.pi * n_days / 7, n_days)) - 20 * np.cos(np.linspace(0, 4 * np.pi * n_days / 7, n_days))

trend = np.linspace(0, 100, n_days)

np.random.seed(42)
noise = np.random.normal(0, 50, n_days)


electricity_consumption = base_consumption + yearly_seasonality + weekly_seasonality + trend + noise


electricity_consumption[electricity_consumption < 0] = 0

df = pd.DataFrame({'Date': dates, 'Consumption': electricity_consumption})
df.set_index('Date', inplace=True)

print("Synthetic Data Head:")
display(df.head())
print("\nSynthetic Data Tail:")
display(df.tail())

plt.figure(figsize=(15, 6))
plt.plot(df.index, df['Consumption'])
plt.title('Synthetic Electricity Consumption Data')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.grid(True)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Consumption'].values.reshape(-1, 1))

print("Scaled Data Shape:", scaled_data.shape)

print("Scaled Data Sample (first 5):\n", scaled_data[:5].flatten())
def create_sequences(data, time_step):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 30
X, y = create_sequences(scaled_data, time_step)

print(f"Shape of X (input sequences): {X.shape}")
print(f"Shape of y (output values): {y.shape}")


X = X.reshape(X.shape[0], X.shape[1], 1)


print(f"Reshaped X for RNN/LSTM: {X.shape}")
train_size = int(len(X) * 0.8)
test_size = len(X) - train_size

X_train, X_test = X[0:train_size,:], X[train_size:len(X),:]
y_train, y_test = y[0:train_size], y[train_size:len(y)]

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

from tensorflow.keras.layers import Input
rnn_model = Sequential()
rnn_model.add(Input(shape=(time_step, 1)))
rnn_model.add(SimpleRNN(50, return_sequences=False))
rnn_model.add(Dense(1))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')

print("RNN Model Summary:")
rnn_model.summary()


print("\nTraining RNN model...")
history_rnn = rnn_model.fit(X_train, y_train, epochs=20, batch_size=64, verbose=1, validation_data=(X_test, y_test))


plt.figure(figsize=(12, 6))
plt.plot(history_rnn.history['loss'], label='RNN Training Loss')
plt.plot(history_rnn.history['val_loss'], label='RNN Validation Loss')
plt.title('RNN Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

from tensorflow.keras.layers import Input
lstm_model = Sequential()
lstm_model.add(Input(shape=(time_step, 1)))
lstm_model.add(LSTM(50, return_sequences=False))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

print("LSTM Model Summary:")
lstm_model.summary()


print("\nTraining LSTM model...")
history_lstm = lstm_model.fit(X_train, y_train, epochs=20, batch_size=64, verbose=1, validation_data=(X_test, y_test))

plt.figure(figsize=(12, 6))
plt.plot(history_lstm.history['loss'], label='LSTM Training Loss')
plt.plot(history_lstm.history['val_loss'], label='LSTM Validation Loss')
plt.title('LSTM Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

forecast_days = 10 * 365


last_sequence = scaled_data[-time_step:].reshape(1, time_step, 1)

rnn_forecast = []
lstm_forecast = []

last_historical_date = df.index[-1]
forecast_dates = pd.date_range(start=last_historical_date + pd.Timedelta(days=1), periods=forecast_days, freq='D')

print(f"Starting forecasting from: {last_historical_date + pd.Timedelta(days=1)}")
print(f"Forecasting for {forecast_days} days until: {forecast_dates[-1]}")

current_rnn_sequence = last_sequence.copy()
for _ in range(forecast_days):

    next_pred_rnn = rnn_model.predict(current_rnn_sequence, verbose=0)[0]
    rnn_forecast.append(next_pred_rnn)

    current_rnn_sequence = np.append(current_rnn_sequence[:, 1:, :], [[next_pred_rnn]], axis=1)

current_lstm_sequence = last_sequence.copy()
for _ in range(forecast_days):

    next_pred_lstm = lstm_model.predict(current_lstm_sequence, verbose=0)[0]
    lstm_forecast.append(next_pred_lstm)


    current_lstm_sequence = np.append(current_lstm_sequence[:, 1:, :], [[next_pred_lstm]], axis=1)

rnn_forecast = scaler.inverse_transform(np.array(rnn_forecast).reshape(-1, 1))
lstm_forecast = scaler.inverse_transform(np.array(lstm_forecast).reshape(-1, 1))

rnn_forecast_df = pd.DataFrame({'Date': forecast_dates, 'RNN_Consumption': rnn_forecast.flatten()})
lstm_forecast_df = pd.DataFrame({'Date': forecast_dates, 'LSTM_Consumption': lstm_forecast.flatten()})

rnn_forecast_df.set_index('Date', inplace=True)
lstm_forecast_df.set_index('Date', inplace=True)

print("RNN Forecast Head:")
display(rnn_forecast_df.head())
print("\nLSTM Forecast Head:")
display(lstm_forecast_df.head())
plt.figure(figsize=(18, 8))
plt.plot(df.index, df['Consumption'], label='Historical Consumption')
plt.plot(rnn_forecast_df.index, rnn_forecast_df['RNN_Consumption'], label='RNN Forecast (Next 10 Years)', linestyle='--')
plt.plot(lstm_forecast_df.index, lstm_forecast_df['LSTM_Consumption'], label='LSTM Forecast (Next 10 Years)', linestyle='-.')

plt.title('Electricity Consumption Forecasting (RNN vs LSTM)')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from tensorflow.keras.layers import Input, GRU
gru_model = Sequential()
gru_model.add(Input(shape=(time_step, 1)))
gru_model.add(GRU(50, return_sequences=False))
gru_model.add(Dense(1))
gru_model.compile(optimizer='adam', loss='mean_squared_error')

print("GRU Model Summary:")
gru_model.summary()


print("\nTraining GRU model...")
history_gru = gru_model.fit(X_train, y_train, epochs=20, batch_size=64, verbose=1, validation_data=(X_test, y_test))


plt.figure(figsize=(12, 6))
plt.plot(history_gru.history['loss'], label='GRU Training Loss')
plt.plot(history_gru.history['val_loss'], label='GRU Validation Loss')
plt.title('GRU Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
gru_forecast = []
current_gru_sequence = last_sequence.copy()
for _ in range(forecast_days):

    next_pred_gru = gru_model.predict(current_gru_sequence, verbose=0)[0]
    gru_forecast.append(next_pred_gru)

    current_gru_sequence = np.append(current_gru_sequence[:, 1:, :], [[next_pred_gru]], axis=1)

gru_forecast = scaler.inverse_transform(np.array(gru_forecast).reshape(-1, 1))
gru_forecast_df = pd.DataFrame({'Date': forecast_dates, 'GRU_Consumption': gru_forecast.flatten()})
gru_forecast_df.set_index('Date', inplace=True)

print("\nGRU Forecast Head:")
display(gru_forecast_df.head())
plt.figure(figsize=(18, 8))
plt.plot(df.index, df['Consumption'], label='Historical Consumption')
plt.plot(rnn_forecast_df.index, rnn_forecast_df['RNN_Consumption'], label='RNN Forecast (Next 10 Years)', linestyle='--')
plt.plot(lstm_forecast_df.index, lstm_forecast_df['LSTM_Consumption'], label='LSTM Forecast (Next 10 Years)', linestyle='-.')
plt.plot(gru_forecast_df.index, gru_forecast_df['GRU_Consumption'], label='GRU Forecast (Next 10 Years)', linestyle=':')

plt.title('Electricity Consumption Forecasting (RNN vs LSTM vs GRU)')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import tensorflow as tf
from tensorflow.keras.layers import MultiHeadAttention, LayerNormalization, Dropout, Input, Dense, GlobalAveragePooling1D
from tensorflow.keras.models import Model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

embed_dim = 64
num_heads = 4
ff_dim = 64
num_transformer_blocks = 2
dropout_rate = 0.1


time_step = 30

n_days = 365 * 10
dates = pd.date_range(start='2010-01-01', periods=n_days, freq='D')

base_consumption = 1000
yearly_seasonality = 300 * np.sin(np.linspace(0, 2 * np.pi * 10, n_days)) + 200 * np.cos(np.linspace(0, 4 * np.pi * 10, n_days))
weekly_seasonality = 50 * np.sin(np.linspace(0, 2 * np.pi * n_days / 7, n_days)) - 20 * np.cos(np.linspace(0, 4 * np.pi * n_days / 7, n_days))
trend = np.linspace(0, 100, n_days)
np.random.seed(42)
noise = np.random.normal(0, 50, n_days)
electricity_consumption = base_consumption + yearly_seasonality + weekly_seasonality + trend + noise
electricity_consumption[electricity_consumption < 0] = 0

df = pd.DataFrame({'Date': dates, 'Consumption': electricity_consumption})
df.set_index('Date', inplace=True)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Consumption'].values.reshape(-1, 1))

def create_sequences(data, time_step):
    X_seq, y_seq = [], []
    for i in range(len(data) - time_step):
        X_seq.append(data[i:(i + time_step), 0])
        y_seq.append(data[i + time_step, 0])
    return np.array(X_seq), np.array(y_seq)

X, y = create_sequences(scaled_data, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)


train_size = int(len(X) * 0.8)
test_size = len(X) - train_size

X_train, X_test = X[0:train_size,:], X[train_size:len(X),:]
y_train, y_test = y[0:train_size], y[train_size:len(y)]


class PositionalEmbedding(tf.keras.layers.Layer):
    def __init__(self, sequence_length, output_dim, **kwargs):
        super().__init__(**kwargs)
        self.sequence_length = sequence_length
        self.position_embedding_layer = tf.keras.layers.Embedding(
            input_dim=sequence_length, output_dim=output_dim
        )

    def call(self, inputs):

        length = tf.shape(inputs)[-2]
        positions = tf.range(start=0, limit=length, delta=1)
        embedded_positions = self.position_embedding_layer(positions)
        return inputs + embedded_positions

    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "sequence_length": self.sequence_length,
                "output_dim": self.position_embedding_layer.output_dim,
            }
        )
        return config

class TransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=dropout_rate, **kwargs):
        super().__init__(**kwargs)
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential(
            [Dense(ff_dim, activation="relu"), Dense(embed_dim),]
        )
        self.layernorm1 = LayerNormalization(epsilon=1e-6)
        self.layernorm2 = LayerNormalization(epsilon=1e-6)
        self.dropout1 = Dropout(rate)
        self.dropout2 = Dropout(rate)

    def call(self, inputs, training=None):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "embed_dim": self.att.key_dim,
                "num_heads": self.att.num_heads,
                "ff_dim": self.ffn.layers[0].units,
                "rate": self.dropout1.rate,
            }
        )
        return config


def build_transformer_model(input_shape, embed_dim, num_heads, ff_dim, num_transformer_blocks):
    inputs = Input(shape=input_shape) # (time_step, 1)


    x = Dense(embed_dim)(inputs)

    x = PositionalEmbedding(input_shape[0], embed_dim)(x)

    for _ in range(num_transformer_blocks):
        x = TransformerBlock(embed_dim, num_heads, ff_dim)(x)


    x = GlobalAveragePooling1D()(x)
    outputs = Dense(1)(x)

    return Model(inputs=inputs, outputs=outputs)


transformer_model = build_transformer_model((time_step, 1), embed_dim, num_heads, ff_dim, num_transformer_blocks)

transformer_model.compile(optimizer='adam', loss='mean_squared_error')

print("\nTime Series Transformer Model Summary:")
transformer_model.summary()

print("\nTraining Time Series Transformer model...")

history_transformer = transformer_model.fit(X_train, y_train, epochs=20, batch_size=64, verbose=1, validation_data=(X_test, y_test))


plt.figure(figsize=(12, 6))
plt.plot(history_transformer.history['loss'], label='Transformer Training Loss')
plt.plot(history_transformer.history['val_loss'], label='Transformer Validation Loss')
plt.title('Time Series Transformer Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()


forecast_days = 10 * 365
last_sequence = scaled_data[-time_step:].reshape(1, time_step, 1)
last_historical_date = df.index[-1]
forecast_dates = pd.date_range(start=last_historical_date + pd.Timedelta(days=1), periods=forecast_days, freq='D') # Redefine for robustness

transformer_forecast = []
current_transformer_sequence = last_sequence.copy()

for _ in range(forecast_days):

    next_pred_transformer = transformer_model.predict(current_transformer_sequence, verbose=0)[0]
    transformer_forecast.append(next_pred_transformer)


    current_transformer_sequence = np.append(current_transformer_sequence[:, 1:, :], [[next_pred_transformer]], axis=1)

transformer_forecast = scaler.inverse_transform(np.array(transformer_forecast).reshape(-1, 1))

transformer_forecast_df = pd.DataFrame({'Date': forecast_dates, 'Transformer_Consumption': transformer_forecast.flatten()})
transformer_forecast_df.set_index('Date', inplace=True)

print("\nTransformer Forecast Head:")
display(transformer_forecast_df.head())



plt.figure(figsize=(18, 8))
plt.plot(df.index, df['Consumption'], label='Historical Consumption')
plt.plot(transformer_forecast_df.index, transformer_forecast_df['Transformer_Consumption'], label='Transformer Forecast (Next 10 Years)', linestyle=':', color='purple')



plt.title('Electricity Consumption Forecasting (Time Series Transformer)')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd

rnn_pred = rnn_model.predict(X_test)
lstm_pred = lstm_model.predict(X_test)
gru_pred = gru_model.predict(X_test)
transformer_pred = transformer_model.predict(X_test)

rnn_pred = scaler.inverse_transform(rnn_pred)
lstm_pred = scaler.inverse_transform(lstm_pred)
gru_pred = scaler.inverse_transform(gru_pred)
transformer_pred = scaler.inverse_transform(transformer_pred)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1,1))

def get_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return [mse, rmse, mae, r2]

results = pd.DataFrame({
    "Model": ["RNN", "LSTM", "GRU", "Transformer"],
    "MSE": [
        get_metrics(y_test_inv, rnn_pred)[0],
        get_metrics(y_test_inv, lstm_pred)[0],
        get_metrics(y_test_inv, gru_pred)[0],
        get_metrics(y_test_inv, transformer_pred)[0]
    ],
    "RMSE": [
        get_metrics(y_test_inv, rnn_pred)[1],
        get_metrics(y_test_inv, lstm_pred)[1],
        get_metrics(y_test_inv, gru_pred)[1],
        get_metrics(y_test_inv, transformer_pred)[1]
    ],
    "MAE": [
        get_metrics(y_test_inv, rnn_pred)[2],
        get_metrics(y_test_inv, lstm_pred)[2],
        get_metrics(y_test_inv, gru_pred)[2],
        get_metrics(y_test_inv, transformer_pred)[2]
    ],
    "R2": [
        get_metrics(y_test_inv, rnn_pred)[3],
        get_metrics(y_test_inv, lstm_pred)[3],
        get_metrics(y_test_inv, gru_pred)[3],
        get_metrics(y_test_inv, transformer_pred)[3]
    ]
})

print(results.sort_values(by="RMSE"))
```
